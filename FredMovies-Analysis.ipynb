```python
# Load packages
import pandas as pd
```

```python
# Load in data

df = pd.read_csv("freddie_movie_final.csv", index_col=0)
df.head()
```

```python
# data summary
df.info()
```

```python
# View ratings of his movies
# Movie title, movie ratings
df[['title','imdb','metascore', 'rotten_tomatoes','awards']]
```

```python
# View years
df[['title', 'Release date (datetime)']]
```

```python
# Fill in the missing dates of given cells manually 
df.at[22, 'Release date'] = 'December 2, 2001'  # Jack and the beanstalk
df.at[25, 'Release date'] = 'November 10, 2016' # Close to the enemy
df.at[29, 'Release date'] = 'April 27, 2018'    # Taskmaster

# Change them to datetime
for i in [22, 25, 29]:
    df.at[i, 'Release date (datetime)'] = pd.to_datetime(df['Release date'][i])
```

```python
# Verify
df[['title', 'Release date (datetime)']]
```

```python
# Change the date to a datetime object. 
# Set the year in a columns

df['Release date (datetime)'] = pd.to_datetime(df['Release date (datetime)'])
df['year'] = df['Release date (datetime)'].dt.year
```

```python
df[['title', 'Release date (datetime)', 'year', 'imdb', 'metascore', 'rotten_tomatoes', 'awards']]
```

```python
# Get unique co-stars
df[['title','Starring']]
```

```python
# df.groupby('Starring').count()
# df.groupby('Starring').nunique()

stars =  df['Starring']
list(stars)
```

```python
# Get out all instances of the actors and actresses 
astros = []
for ss in list(stars):
    if not isinstance(ss, float):
        actors = ss.strip('[]').split(',')
        for actor in actors:
            actor = actor.replace("'","").strip()
            astros.append(actor)
astros    
```

```python
# count each time an actor's name appears
from collections import Counter
```

```python
Counter(astros)
```

```python
import matplotlib.pyplot as plt
```

```python
# plt.plot(df.groupby(['year']))

results = df.groupby(['year']).count()

years = [year for year, title in  df.groupby('year')]

plt.bar(years, results['title'])
plt.xticks(range(2001,2022), rotation='vertical')
plt.ylabel('Number of Movies/TV shows')
plt.xlabel('Years')
plt.show()
```
